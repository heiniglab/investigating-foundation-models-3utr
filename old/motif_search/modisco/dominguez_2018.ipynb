```python
import pandas as pd
import numpy as np

from collections import defaultdict
import matplotlib.pyplot as plt
```

```python
motif_len = 6 #5 or 6
```

```python
data_dir = '/lustre/groups/epigenereg01/workspace/projects/vale/MLM/'
```

```python
table_motifs = pd.read_excel(data_dir + 'dominguez_2018/1-s2.0-S1097276518303514-mmc4.xlsx',
             sheet_name = motif_len-4, skiprows=1)
```

```python
motifs_PWM = {}
motifs_data = {}

for col_idx in range(0,len(table_motifs.columns),2):
    
    protein_name = table_motifs.columns[col_idx]
    X = table_motifs.iloc[:,col_idx:col_idx+2]
    PWM = PWM_from_weights(X)
    PWM = pd.DataFrame(PWM,columns=["A", "C", "G", "T"])
    motifs_PWM[protein_name] = PWM
    
    protein_df = table_motifs.iloc[:,col_idx:col_idx+2]
    protein_df.columns = ['motif','stepwise_r']
    protein_df = protein_df[~protein_df['motif'].isna()]
    motifs_data[protein_name] = protein_df

motifs_PWM = pd.concat(motifs_PWM)
motifs_data = pd.concat(motifs_data)
```

```python
def PWM_from_weights(X):
    
    motifs = []
    weights = []
    
    for motif, weight in X.values:
        if  np.isnan(weight):
            break
        motifs.append(list(motif))
        weights.append(weight)
    
    motifs = np.array(motifs)
    weights = np.array(weights)

    #print(motifs)
    #print(weights)
    
    PWM = np.zeros((motif_len,4))
    
    for pos_idx in range(motif_len):
        for base_idx,base in enumerate('ACGT'):
           PWM[pos_idx,base_idx] = weights[motifs[:,pos_idx]==base].sum()

    return PWM
```

```python
import logomaker
```

```python
selected_proteins = ['EWSR1', 'HNRNPL', 'PABPN1L', 'KHSRP', 'RBFOX2', 'HNRNPC', 'PUM1', 'FUS', 'TAF15',
                     'TRA2A', 'PCBP2', 'TARDBP', 'TIA1', 'PTBP3', 'CELF1', 'FUBP3', 'KHDRBS2'] #top motifs overlap for RBNS and eCLIP


N_prot = len(selected_proteins)

fig, axes = plt.subplots(N_prot,1, figsize=(2,15),dpi=300)

axes = axes.flatten()

for c, protein_name in enumerate(selected_proteins):

    PWM = motifs_PWM.loc[protein_name]

    PWM = logomaker.transform_matrix(PWM, from_type='probability', to_type='information')

    ax = axes[c]

    logomaker.Logo(PWM,ax = ax)

    ax.set_title(protein_name)
    ax.set_xticks([])
    ax.set_ylim(min(PWM.sum(axis=1).min(), 0), PWM.sum(axis=1).max())

    if c==N_prot-1:
        break

fig.tight_layout()
```
